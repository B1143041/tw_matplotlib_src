```python
import pandas as pd
```

```python
df=pd.read_csv("scores/gradebook-IE-2020-2.csv")
```

```python
# replace " (OT)" and delete the average
df=df.iloc[:-1,:].copy()
```

```python
df.fillna(0,inplace=True)
```

```python
df.replace('-',0,inplace=True)
```

```python
q_cols=["First Term Quiz, Sequences and Series 100 pts", "First Term Quiz, re-opened 100 pts"]
for q_col in q_cols:
    df[q_col]=pd.to_numeric(df[q_col]).values
```

```python
df['quiz1']=round(df[q_cols].max(axis=1))
```

```python
df[["First Term Quiz, Sequences and Series 100 pts", "First Term Quiz, re-opened 100 pts"]].max()
```

```python
df['quiz2']=df[['Term Quiz-2 100 pts']]
```

```python
num_columns=['Series I 100 pts','Series 2, Other Tests 100 pts',\
             'Taylor Series 3 100 pts','Partial Differentiation, Practices 100 pts', \
             'Optimization-1 100 pts','optimization-2_Lagrange_Multiplier 100 pts']
name='p'
i=1
for col in num_columns:
    df[col]=pd.to_numeric(df[col]).values
  
```

```python
df['practices']=df[num_columns].mean(axis=1).values
```

```python
df['practices']=round((df['practices']/df['practices'].max())*100)
```

```python
df[['quiz1','practices']].plot(kind='hist',figsize=[12,6],alpha=0.3,title='Histogram of Score of IE108')
```

```python
df.sort_values(by=['Username'])[['Username','quiz1','practices']].to_csv("scores/IE108-2.csv",index=False)
```
